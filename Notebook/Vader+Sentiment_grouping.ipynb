```python
import pandas as pd
```

```python
df = pd.read_csv("Tweets_with_sentiment.csv")
```

```python
df
```

```python
df = df[~df['cleaned_text'].isna()].reset_index(drop=True)
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
def compound_sentiment(text):
    sentiment_score = analyzer.polarity_scores(text)
    return sentiment_score['compound']  
df['Sentiment_vader'] = df['cleaned_text'].apply(compound_sentiment)
```

```python
df["kk08_score_diff"] = df["Positive_score"] - (1-df["Positive_score"])
```

```python
df = df.rename(columns={"Positive_score":"kk08_pos_score"})
```

```python
df
```

```python
def pred(x):
    if "predict" in x.lower():
        return 1
    else:
        return 0
    
def analysis(x):
    if "analys" in x.lower():
        return 1
    else:
        return 0   

def short(x):
    if "short" in x.lower():
        return 1
    else:
        return 0

def long(x):
    if "long" in x.lower():
        return 1
    else:
        return 0
    
def buy(x):
    if "buy" in x.lower() or "bought" in x.lower():
        return 1
    else:
        return 0
    
def sell(x):
    if "sell" in x.lower() or "sold" in x.lower():
        return 1
    else:
        return 0
    

def bull(x):
    if "bull" in x.lower():
        return 1
    else:
        return 0

def bear(x):
    if "bear" in x.lower():
        return 1
    else:
        return 0
    
def bankrupt(x):
    if "bankrupt" in x.lower():
        return 1
    else:
        return 0
    
def fraud(x):
    if "fraud" in x.lower():
        return 1
    else:
        return 0
    
# central banks digital currency
def cbdc(x):
    if "cbdc" in x.lower():
        return 1
    else:
        return 0
    
def record(x):
    if "record" in x.lower():
        return 1
    else:
        return 0
    
def crisis(x):
    if "crisis" in x.lower():
        return 1
    else:
        return 0
    
def collapse(x):
    if "collapse" in x.lower():
        return 1
    else:
        return 0
    
# all times high
def ATH(x):
    if "ath" in x.lower():
        return 1
    else:
        return 0

def grow(x):
    if "grow" in x.lower() or "grew" in x.lower():
        return 1
    else:
        return 0
    
def rise(x):
    if "rise" in x.lower() or "rose" in x.lower() or "risen" in x.lower():
        return 1
    else:
        return 0
    
def fall(x):
    if "fall" in x.lower() or "fell" in x.lower():
        return 1
    else:
        return 0
```

```python
df["pred"] = df["text"].apply(pred)
df["analysis"] = df["text"].apply(analysis)
df["short"] = df["text"].apply(short)
df["long"] = df["text"].apply(long)
df["buy"] = df["text"].apply(buy)
df["sell"] = df["text"].apply(sell)
df["bull"] = df["text"].apply(bull)
df["bear"] = df["text"].apply(bear)
df["bankrupt"] = df["text"].apply(bankrupt)
df["fraud"] = df["text"].apply(fraud)
df["cbdc"] = df["text"].apply(cbdc)
df["record"] = df["text"].apply(record)
df["crisis"] = df["text"].apply(crisis)
df["collapse"] = df["text"].apply(collapse)
df["ATH"] = df["text"].apply(ATH)
df["grow"] = df["text"].apply(grow)
df["rise"] = df["text"].apply(rise)
df["fall"] = df["text"].apply(fall)

```
