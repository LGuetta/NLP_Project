```python
import pandas as pd
```

```python
df = pd.read_csv("Binance_BTCUSDT_2022_minute.csv")
```

```python
len(df)
# full year?
```

```python
df_month = df[(df["date"].str.contains("2022-01")) | (df["date"].str.contains("2022-02-01"))].reset_index(drop=True)
# one more day, because our sentiment is lagged
```

```python
df_month["date"] = pd.to_datetime(df_month["date"])
```

```python
df_month
```

```python

df_month.set_index('date', inplace=True)
# 15 minutes
df_15m = df_month.resample('15min').agg({'open': 'first', 'close': 'last'}).reset_index()
df_15m['Return'] = (df_15m['close'] - df_15m['open']) / df_15m['open']

# 30 minutes
df_30m = df_month.resample('30min').agg({'open': 'first', 'close': 'last'}).reset_index()
df_30m['Return'] = (df_30m['close'] - df_30m['open']) / df_30m['open']

df_month = df_month.reset_index()


```

```python
df_month.iloc[-15]
```

```python
def sign(x):
    return 1 if x>=0 else 0
```

```python
df_15m
```

```python
df_15m["Return_sign"] = df_15m["Return"].apply(sign)
```

```python
df_15m
```

```python
df_15m.to_csv("BTCUSDT_15m.csv",index=False)
```

```python
df_30m["Return_sign"] = df_30m["Return"].apply(sign)
```
