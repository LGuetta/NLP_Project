# 15 min

```python
import pandas as pd
```

```python
sentiment_15m = pd.read_csv("Tweets_with_sentiment_15m_extra.csv")
```

```python
fin_15m = pd.read_csv("BTCUSDT_15m.csv")
```

```python
sentiment_15m
```

```python
fin_15m
```

```python
df_15m = pd.merge(sentiment_15m, fin_15m, left_on = "Date_15m", right_on="date", how = "inner")
```

```python
df_15m
```

```python
df_15m.to_csv("Full_data_15m.csv", index = False)
```

# 1 hour

```python
sentiment_1H = pd.read_csv("Tweets_with_sentiment_1H_extra.csv")
fin_1H = pd.read_csv("BTCUSDT_1H.csv")
```

```python
sentiment_1H
```

```python
fin_1H
```

```python
sentiment_1H = sentiment_1H.rename(columns={"Date_hourly":"Date_1H"})
fin_1H = fin_1H.rename(columns={"Date":"date", "Open":"open", "Close":"close"})[["date", "open", "close","Return","Return_sign"]]
```

```python
fin_1H
```

```python
df_1H = pd.merge(sentiment_1H, fin_1H, left_on = "Date_1H", right_on="date", how = "inner")
df_1H
```

```python
df_1H.to_csv("Full_data_1H.csv", index = False)
```
